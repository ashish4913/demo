**1.import required liberary**

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**2.Load dataset**

```
df=pd.read_csv("health care diabetes.csv")
df
df1=df.copy()
```

**7 check missing values**

```
df.isnull().sum()
```

There is no missing value in our dataset

**8 check missing values using heat map**

```
plt.figure(figsize=(5,3),dpi=100)
plt.title('Checking Missing Value with Heatmap')
sns.heatmap(df.isnull(),cmap='magma',yticklabels=False)
```

**9.check for zeros in dataset**

```
df.isna().sum()
```

**10.Repace zero with NaN**

```
df.replace(0, np.nan, inplace=True)
df
```

**11.Replace NaN value accoring to datafield**

```
df['Pregnancies'].fillna(df['Pregnancies'].value_counts().index[0],inplace=True)
df['Glucose'].fillna(df['Glucose'].mean(),inplace=True)
df['BloodPressure'].fillna(df['BloodPressure'].mean(),inplace=True)
df['SkinThickness'].fillna(df['SkinThickness'].mean(),inplace=True)
df['Insulin'].fillna(df['Insulin'].mean(),inplace=True)
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
df['Outcome'].fillna(df['Outcome'].value_counts().index[0],inplace=True)
```

**12.Again check for NaN value**

```
df.isna().sum()
```

No missing value

**13 Analysis using histogram**

```
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
sns.distplot(df['Pregnancies'],kde=True,ax=axes[0,0])
sns.distplot(df['Glucose'],kde=True,ax=axes[0,1])
sns.distplot(df['BloodPressure'],kde=True,ax=axes[0,2])
sns.distplot(df['SkinThickness'],kde=True,ax=axes[0,3])
sns.distplot(df['Insulin'],kde=True,ax=axes[1,0])
sns.distplot(df['BMI'],kde=True,ax=axes[1,1])
sns.distplot(df['DiabetesPedigreeFunction'],kde=True,ax=axes[1,2])
sns.distplot(df['Age'],kde=True,ax=axes[1,3])
```

**16.frequency plot for each feature with outcome**

**17 Create scatter charts between the pair of variables to understand the relationships**

```
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
sns.scatterplot(x='Pregnancies',y='BloodPressure',data=df1,ax=axes[0,0])
sns.scatterplot(x='Glucose',y='BloodPressure',data=df1,ax=axes[0,1])
sns.scatterplot(x='BMI',y='BloodPressure',data=df1,ax=axes[0,2])
sns.scatterplot(x='SkinThickness',y='Age',data=df1,ax=axes[0,3])
sns.scatterplot(x='BMI',y='Age',data=df1,ax=axes[1,0])
sns.scatterplot(x='BloodPressure',y='Outcome',hue="Age",data=df1,ax=axes[1,1])
sns.scatterplot(x='BloodPressure',y='DiabetesPedigreeFunction',data=df1,ax=axes[1,2])
sns.scatterplot(x='Insulin',y='Age',data=df1,ax=axes[1,3])
```
